Handeling Missing Values in numpy

```python
# np.isnan(array)
import numpy as np 
arr=np.array([3,8,6,np.nan,67,90,np.nan,100])
print(np.isnan(arr))
```

```python
# np.nan_to_num(array,nan=value)
a=np.nan_to_num(arr,nan=10)
print(a)
```

```python
# Infinty
# np.isinf(array)
l=np.array([56,89,20,np.inf,78,-np.inf])
result=np.isinf(l)
print(result)
# solve(prossece one)
# l[np.isinf(l)] = 0
# print(l)
r=np.nan_to_num(l,posinf=6,neginf=0)
print(r)
```
